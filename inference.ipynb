```python
import os
import roboflow
```

```python
rf = roboflow.Roboflow(api_key=os.environ['ROBOFLOW_API_KEY'])
```

```python
project = rf.workspace().project('hts-uki6g')
model = project.version('1').model
```

```python

# optionally, change the confidence and overlap thresholds
# values are percentages
model.confidence = 25
model.overlap = 0

# predict on a local image
image_path = os.path.join(os.environ['ROOT_FOLDER'], 'data/images/schools/Akins HS Building 2.png')
prediction = model.predict(image_path)
```

```python

# Plot the prediction in an interactive environment
prediction.plot()
```

```python

# Convert predictions to JSON
prediction.json()
```
